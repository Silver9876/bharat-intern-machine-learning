HOUSE PRICE PREDICTION MODEL

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
#importing the .csv file into our model
data=pd.read_csv("USA_Housing.csv")
data.head()
```

```python
data.dropna(inplace=True) #removing the missing values from the data file for the model
data.info()
```

```python
#we are using multiple features to estimate the price of the house
X=data[['Avg. Area House Age','Avg. Area Number of Rooms','Avg. Area Number of Bedrooms','Avg. Area Number of Bedrooms','Area Population']]
y=data['Price']
```

```python
#spliting the data into training and testing data sets using the python sklearn library
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(X, y, test_size=0.6)
```

MODEL EVALUATION

```python
#Importing the linear regression module from the sklearn
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
#training the model using the training data set
reg.fit(x_train, y_train)
```

```python
#predicting the price of the house using the testing data set
y_pred=reg.predict(x_test)
```

```python
plt.scatter(y_test, y_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='-', color='black', label='Ideal Line')
plt.title('Scatter plot of price vs various features of a house')
plt.xlabel('features')
plt.ylabel('price (dollars)')
plt.show()
```

```python
from sklearn import metrics
print("mean_squared_error:",metrics.mean_squared_error(y_test, y_pred))
```
